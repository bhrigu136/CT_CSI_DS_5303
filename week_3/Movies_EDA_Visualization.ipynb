```python
# TMDb Movies Dataset - Exploratory Data Analysis & Visualization
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set visual style
sns.set(style="whitegrid")
%matplotlib inline
```

```python
df = pd.read_csv("tmdb_5000_movies.csv")
```

```python
print("First 5 rows of the dataset:")
display(df.head())
```

```python
print("\nDataset Info:")
df.info()
```

```python
print("\nSummary Statistics:")
display(df.describe())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
## Data Visualization
```

```python
# 1. Distribution of Popularity
plt.figure(figsize=(8,5))
sns.histplot(df['popularity'], bins=30, kde=True)
plt.title('Distribution of Movie Popularity')
plt.xlabel('Popularity Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Budget vs Revenue Scatter Plot
plt.figure(figsize=(10,6))
sns.scatterplot(x='budget', y='revenue', data=df)
plt.title('Budget vs Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

```python
# 3. Top 10 Most Popular Movies
top_popular = df.sort_values(by='popularity', ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x='popularity', y='original_title', data=top_popular)
plt.title('Top 10 Most Popular Movies')
plt.xlabel('Popularity')
plt.ylabel('Movie Title')
plt.show()
```

```python
# 4. Top 10 Revenue-Generating Movies
top_revenue = df.sort_values(by='revenue', ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x='revenue', y='original_title', data=top_revenue)
plt.title('Top 10 Highest Revenue Movies')
plt.xlabel('Revenue')
plt.ylabel('Movie Title')
plt.show()
```

```python
# 5. Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```
