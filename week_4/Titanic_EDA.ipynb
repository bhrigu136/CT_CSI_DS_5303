# Titanic Dataset - In-depth Exploratory Data Analysis (EDA)

```python
## Step 1: Import Required Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
%matplotlib inline
```

```python
## Step 2: Load the Dataset
df = pd.read_csv("train.csv")
```

```python
## Step 3: Basic Exploration
df.info()
print("\nSummary Statistics:")
print(df.describe(include='all'))
print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())
```

```python
## Step 4: Handling Missing Values
# Fixing chained assignment warning by using assignment instead of inplace
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.drop('Cabin', axis=1, errors='ignore')
```

```python
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

```python
## Step 5: Data Distributions
```

```python
# Age distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Fare distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
## Step 6: Box Plots (Detecting Outliers)
```

```python
# Boxplot of Fare
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Fare'])
plt.title('Boxplot of Fare')
plt.show()
```

```python
# Boxplot of Age
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Age'])
plt.title('Boxplot of Age')
plt.show()
```

```python
## Step 7: Correlation Heatmap
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```
